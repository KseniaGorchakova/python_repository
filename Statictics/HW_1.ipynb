### Исследование наличия выбросов в столбце pulse

```python
# Выбросы в столбце pulse
q1 = df['pulse'].quantile(0.25)
q3 = df['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['pulse'].between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')
remove_outliers
```

```python
pd.concat([df, remove_outliers]).drop_duplicates(keep=False)
```

```python
remove_outliers.describe()
```

#### Я бы оставила даные выбросы, т.к. таких кейсов немного (5 штук) и возможно это реальные данные по пульсу лошадей. Медиана остается той же, другие статистики тоже не сильно изменяются или вообще не меняются, значение СКО при удалении этих данных снижается до 26,08

### Замена пустых значений в столбце surgery? на моду, т.к. это категориальная переменная и логично заменить на наиболее часто встречающуеся значение
