```python
import numpy as np
from scipy.misc import derivative 
import matplotlib.pyplot as plt
from sympy import *
```

### Уровень 0: 
#### Посчитайте значение первой и второй производной функции:
#### f(x) = x^5 + 4sin(2x) + cos(3x+3)
#### В точке x = 1.

```python
x = symbols('x')
```

```python
def f(x):
    return x**5 + 4*np.sin(2*x) + np.cos(3*x+3)
```

```python
# Первая производная
derivative(f, 1, dx=1e-6)
```

```python
# Вторая производная
derivative(f, 1, n=2, dx=1e-6)
```

```python
# Первая и вторая производные функции
exp = x**5 + 4*sin(2*x) + cos(3*x+3)
print("Before Differentiation : {}".format(exp))
  
# Use sympy.diff() method
dif = diff(exp, x)
  
print("First Differentiation : {}".format(dif))

dif_2 = diff(diff(exp, x),x)
  
print("Second Differentiation : {}".format(dif_2))
```

### Уровень 1:
#### Постройте график первой и второй производной функции:
#### f(x) = sin^5(2x+1)
#### На отрезке [-5; 5].

```python
x_1 = symbols('x_1')
```

```python
def f_1(x_1):
    return np.sin(2*x_1+1)**5
```

```python
x_1 = np.linspace( -5 , 5, 100)
```

```python
# Первая производная
fx = f_1(x_1)
f1x = [derivative(f_1, xx, n =1) for xx in x_1]

# Вторая производная
f2x = [derivative(f_1, xx, n =2) for xx in x_1]
```

```python
# График первой производной
plt.plot(x_1, fx, label = 'function' )
plt.plot(x_1, f1x, label = 'first derivative' )
plt.legend()
plt.show()
```

```python
# График второй производной
plt.plot(x_1, fx, label = 'function' )
plt.plot(x_1, f2x, label = 'second derivative' )
plt.legend()
plt.show()
```
