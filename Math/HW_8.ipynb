```python
import scipy.stats as sts
import matplotlib.pyplot as plt
import numpy as np
import math
```

### Задание №1

### Вариант 1

```python
# Нормальное распределение
mu = 2
sigma = 0.5

norm_rv = sts.norm(loc = mu, scale = sigma)

norm_rv.rvs(size = 10)
```

```python
# Экспоненциальное распределение
lam = 0.1
expron_rv = sts.expon(lam)
```

```python
# Распределение Стьюдента
n = 10
t_rv = sts.t(n)
t_rv.rvs(10)
```

```python
x = np.linspace(-5, 5, 100)
pdf = norm_rv.pdf(x)
pdf_1 = expron_rv.pdf(x)
pdf_2 = t_rv.pdf(x)
plt.plot(x, pdf)
plt.plot(x, pdf_1)
plt.plot(x, pdf_2)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

### Задание 3
Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?

```python
total = red + yel + gr # общее время светофора для всех цветов
red = 54
yel = 4
gr = 28
```

```python
p_1 = red / total
p_2 = yel / total
p_3 = gr / total
print(p_1)
print(p_2)
print(p_3)
```

```python
I = - (p_1 * math.log2(p_1) + p_2 * math.log2(p_2) + p_3 * math.log2(p_3))
print(I, "бит")
```
